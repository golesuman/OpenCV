# Canny Edge Detection
Widely used and accurate edge detection algorithm

1. Filter out noise using a Gaussian Blur
2. Find the strength and direction of edges using sobel filters

3. Applies non-maximum supression to isolate the strongest images and thin them to one-pixel wide lines

4. Uses hysteresis to isolate the best edges

```python
import numpy as np
import cv2 
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('sunflower.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
gray_img=np.copy(img)
gray_img=cv2.cvtColor(gray_img,cv2.COLOR_RGB2GRAY)
plt.imshow(gray_img,cmap='gray')
```

```python
#canny edge detection
#threshold e
lower = 120
upper = 240
edges = cv2.Canny(gray_img, lower, upper)
plt.imshow(edges, cmap='gray')
```

```python
edges1 = cv2.Canny(gray_img, 30 , 100)
edges2 = cv2.Canny(gray_img, 100, 240)
fig, (ax1, ax2) = plt.subplots(1,2, figsize = (10, 8))
ax1.imshow(edges1, cmap='gray')
ax2.imshow(edges2, cmap='gray')
```
