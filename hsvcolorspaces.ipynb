```python
import cv2 
import matplotlib.pyplot as plt
import numpy as np
```

```python
img=cv2.imread('water_balloons.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
img_copy=np.copy(img)
plt.imshow(img)
```

```python
r=img_copy[:,:,0]
g=img_copy[:,:,1]
b=img_copy[:,:,2]

f,(ax1,ax2,ax3) = plt.subplots(1,3,figsize=(20,10))
ax1.set_title('Red')
ax1.imshow(r,cmap='gray')

ax2.set_title('Green')
ax2.imshow(g,cmap='gray')

ax3.set_title('Blue')
ax3.imshow(b,cmap='gray')
```

```python
hsv=cv2.cvtColor(img_copy,cv2.COLOR_RGB2HSV)
plt.imshow(hsv)
```

```python
h=hsv[:,:,0]
s=hsv[:,:,1]
v=hsv[:,:,2]

f,(ax1,ax2,ax3) = plt.subplots(1,3,figsize=(20,10))
ax1.set_title('Hue')
ax1.imshow(h,cmap='gray')

ax2.set_title('Saturation')
ax2.imshow(s,cmap='gray')

ax3.set_title('Value')
ax3.imshow(v,cmap='gray')
```

```python
lower_pink=np.array([180,0,110])
higher_pink=np.array([255,255,225])
rgb_map=cv2.inRange(img_copy,lower_pink,higher_pink)
plt.imshow(rgb_map,cmap='gray')
```

```python
masked_img=np.copy(img_copy)
masked_img[rgb_map==0]=[0,0,0]
plt.imshow(masked_img)
```

```python
lower_hue=np.array([160,0,0])
higher_hue=np.array([180,255,255])
hue_mask=cv2.inRange(hsv,lower_hue,higher_hue)
plt.imshow(hue_mask,cmap='gray')
```

```python
hue_maskedimg=np.copy(img_copy)
hue_maskedimg[hue_mask==0]=[0,0,0]
plt.imshow(hue_maskedimg)
```
