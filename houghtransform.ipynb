```python
import cv2 
import matplotlib.pyplot as plt
import numpy as np
```

```python
img=cv2.imread('phone.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

Peform Edge detection

```python
lower=50
higher=100
gray_img=np.copy(img)
gray_img = cv2.cvtColor(gray_img, cv2.COLOR_RGB2GRAY)
plt.imshow(gray_img,cmap='gray')
```

```python
canny=cv2.Canny(gray_img,lower,higher)
plt.imshow(canny,cmap='gray')
```

Hough transformation


```python
rho = 1
theta = np.pi/180
threshold = 60
min_line_length = 100
max_line_gap = 5

lines = cv2.HoughLinesP(canny, rho, theta, threshold, np.array([]),
                        min_line_length, max_line_gap)
lineImg = np.copy(img)
for line in lines:
  for x1,y1,x2,y2 in line:
    cv2.line(lineImg,(x1,y1), (x2,y2), (255,0,0),5)
plt.imshow(lineImg)
```
