## What are low pass fiters ? 
Low-pass filters are those filters which are used for removing the noise from an image. Noises can be pretty daunting while processing the image and they can reduce our accuracy and effeciency of model. This step is done before passing the image into any high-pass filters.

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img=cv2.imread('brainimg.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
img_copy=np.copy(img)
gray_img=cv2.cvtColor(img_copy,cv2.COLOR_RGB2GRAY)
plt.imshow(gray_img,cmap='gray')
```

Averaging Filter is one example of low-pass filters

lowpassfilters.png


```python
#gaussian blur -- low pass filter uses 
gaussian_img = cv2.GaussianBlur(gray_img,(5,5),0)
fig, axes = plt.subplots(1,2,figsize=(20,8))
axes[0].set_title("Normal")
axes[0].imshow(gray_img,cmap='gray')
axes[1].set_title('Blurred Img')
axes[1].imshow(gaussian_img,cmap='gray')
```

```python
filter=np.array([[-2, 0, 2],
                 [-1, 0, 1],
                 [-2, 0, 2]])
filtered=cv2.filter2D(gray_img,-1,filter)
filtered_blurred=cv2.filter2D(gaussian_img,-1,filter)
fig,axes = plt.subplots(1,2,figsize=(20,8))
axes[0].set_title("Original")
axes[0].imshow(filtered,cmap='gray')
axes[1].set_title('Blurred Img')
axes[1].imshow(filtered_blurred,cmap='gray')
```
