```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img=cv2.imread('cityhall.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

plt.imshow(img)
```

```python
img_copy=np.copy(img)
gray_img=cv2.cvtColor(img_copy,cv2.COLOR_RGB2GRAY)
plt.imshow(gray_img,cmap='gray')
```

```python
kernel=np.array([[-1,0,1],
                 [-2,0,2],
                 [-1,0,1]])
filter_img=cv2.filter2D(gray_img,-1,kernel)
fig,axes=plt.subplots(2,1,figsize=(10,10))
axes[0].imshow(kernel,cmap='gray')
axes[0].set_title('Filter')
axes[1].imshow(filter_img,cmap='gray')
axes[1].set_title('Filtered Image')

```

```python
#creating binary Image
retval, binary_img=cv2.threshold(filter_img,100,255,cv2.THRESH_BINARY)
```

```python
plt.imshow(binary_img,cmap='gray')
```
